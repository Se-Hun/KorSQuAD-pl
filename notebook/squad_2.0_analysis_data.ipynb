```python
data_dir = "../data/"
data_name = "squad_v2.0"

fns = {
    "train" : os.path.join(data_dir, data_name, "train.json"),
    "test" : os.path.join(data_dir, data_name, "dev.json")
}
```

```python
train_fn = fns["train"]
test_fn = fns["test"]

train_data = None
test_data = None
with open(train_fn, 'r', encoding='utf-8') as f:
    train_data = json.load(f)["data"]
    
with open(test_fn, 'r', encoding='utf-8') as f:
    test_data = json.load(f)["data"]
```
